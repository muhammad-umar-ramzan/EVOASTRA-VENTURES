```python
from bs4 import BeautifulSoup
import math
import pandas as pd

# Load the HTML file
with open('hyundai.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
car_divs = soup.find_all('div', class_='_2YB7p')
```

```python
car_divs
```

```python
car_data = []
```

```python
for car in car_divs:
    car_full_name = car.find('h3', class_='_11dVb').text
    year, car_name = car_full_name.split(' ', 1)
    details = car.find('ul', class_='_3J2G-').find_all('li')
    km_driven = details[0].text
    fuel_type = details[2].text
    transmission = details[4].text
    price = car.find('strong', class_='_3RL-I').text.strip('₹').replace(' Lakh', '').strip()
    price_in_inr = float(price) * 100000
    location = car.find('p', class_='_3dGMY').find_all('span')[-1].text.strip()
    
    # Append car details to the list
    car_data.append([year, car_name, km_driven, fuel_type, transmission, location, math.floor(price_in_inr)])
```

```python
df = pd.DataFrame(car_data, columns=['Year', 'Car_Name', 'Km_Driven', 'Fuel_Type', 'Transmission', 'Location', 'Price'])
```
